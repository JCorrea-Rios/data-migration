```python
def get_matches_df_name_vector_dimensionality(sparse_matrix, name_vector, top=100):
    non_zeros = sparse_matrix.nonzero()
    
    sparserows = non_zeros[0]
    sparsecols = non_zeros[1]
    
    if top:
        nr_matches = top
    else:
        nr_matches = sparsecols.size
    
    matches_dict = {name: [] for name in name_vector}
    
    for index in range(0, nr_matches):
        left_name = name_vector[sparserows[index]]
        right_name = name_vector[sparsecols[index]]
        similarity = sparse_matrix.data[index]

        matches_dict[left_name].append(right_name)
    
    return matches_dict
```
